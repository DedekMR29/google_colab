```python
# name file : Perceptiron output tunggal dengan Keras
# library numpy : fungsi array untuk merubah list menjadi array
# library Keras.model  : fungsi sequential()
#                      : Fungsi add() untuk menambahkan layer ke model
# library keras.layers : fungsi dense() untuk membuat layer dense pada model
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

#menyiapkan Data Set
#Input dan Target T
#langkah 1: menyiapkan data set input dan output
X = np.array([[0,0],
             [0,1],
             [1,0],
             [1,1]])
T = np.array([[1],
              [0],
              [0],
              [0]])

# mendefinisikan model JST dengan outut 1 dan input 2
# dalam jst kita terapkan fungsi sigmoid 
# pada note pada jst tersebut fi tambahkan bias 
# sesuai dengan arsitektur yaitu preptron
# dengan vektor input 2 dan vektor output 1 maka
# jumlah node pada layer adalah sesuai dengan jumlah vektor output

#langkah 2, Mendefinisikan model JST
model = Sequential()
#menambahkan Layer Dense ke model
NNode = 1 #Jumlah Node
NInput = 2 #Jumlah vektor input
model.add(Dense(NNode, input_dim=NInput, activation='sigmoid', use_bias=True))

#3. Mengcompile model
#parameter Training
#fungsi loss = mse
#optimizer : SGD Stochastic Gradien Descent
#metrics : accuracy
model.compile(loss='mse', optimizer='SGD',metrics=['accuracy'])

#4. Training Model menggunakan 4000 epoh
#X : Data Input
# T : Target
#epoh : Jumlah iterasi = 4000
His = model.fit(X, T, epochs=4000)

#menampilkan Grafik loss
plt.plot(His.history['loss'])


#5. Prediksi
yp = model.predict(X)
#mencetak hasil prediksi
print(yp)
```

```python
# name file : Perceptiron output tunggal dengan Keras
# library nuumpy : fungsi array untuk merubah list menjadi array
# library Keras.model  : fungsi sequential()
#                      : Fungsi add() untuk menambahkan layer ke model
# library keras.layers : fungsi dense() untuk membuat layer dense pada model
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt

#menyiapkan Data Set
#Input dan Target T
#langkah 1: menyiapkan data set input dan output
X = np.array([[0,0],
             [0,1],
             [1,0],
             [1,1]])
T = np.array([[1],
              [0],
              [0],
              [0]])

# mendefinisikan model JST dengan outut 1 dan input 2
# dalam jst kita terapkan fungsi sigmoid 
# pada note pada jst tersebut di tambahkan bias 
#jumlah hidden layer 3 dengan masing-masing node 1000
#unit aktivasi relu kecuali node terakhir sigmoid untuk membatasi 
# output diantara 0 dan 1

#langkah 2, Mendefinisikan model JST
model = Sequential()
#menambahkan Layer Dense ke model
NOoutput = 1 #Jumlah Node
NInput = 2 #Jumlah vektor input
model.add(Dense(1000, input_dim=NInput, activation='relu', use_bias=True))
model.add(Dense(1000, activation='relu', use_bias=True))
model.add(Dense(NOoutput, activation='sigmoid', use_bias=True))

#3. Mengcompile model
#parameter Training
#fungsi loss = mse
#optimizer : SGD Stochastic Gradien Descent
#metrics : accuracy
model.compile(loss='mse', optimizer='SGD',metrics=['accuracy'])

#4. Training Model menggunakan 4000 epoh
#X : Data Input
# T : Target
#epoh : Jumlah iterasi = 4000
His = model.fit(X, T, epochs=100)

#menampilkan Grafik loss
plt.plot(His.history['loss'])


#5. Prediksi
yp = model.predict(X)
#mencetak hasil prediksi
print(yp)
```
